```python
import unitvelo as utv
```

```python
velo = utv.config.Configuration()
velo.R2_ADJUST = True 
velo.IROOT = None
velo.FIT_OPTION = '1'
velo.GPU = -1
velo.BASIS = 'pca'
```

```python
import scvelo as scv
adata_toggle = scv.read('../data/toggle_switch/data_toggle.h5ad')
```

```python
adata_toggle.var['highly_variable']=True
```

```python
adata_toggle.obsm['X_coord'] = adata_toggle.obsm['coord']
```

```python
adata_toggle.write('../data/toggle_switch/data_toggle.h5ad')
```

```python
path_to_adata = '../data/toggle_switch/data_toggle.h5ad'
label = 'leiden'
adata = utv.run_model(path_to_adata, label, config_file=velo)
```

```python
scv.pl.velocity_embedding_stream(adata, color='leiden', basis = 'coord', dpi=100, title='')
```
